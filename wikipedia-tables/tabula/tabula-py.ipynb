```python
#pdf_path = "https://github.com/chezou/tabula-py/raw/master/tests/resources/data.pdf"

pdf_path = "/Users/travishartman/Desktop/wiki/pdf/pdf_to_table/pdfs/tef.pdf"

dfs = tabula.read_pdf(pdf_path, pages="all", stream=False)
# read_pdf returns list of DataFrames
print(len(dfs))

for df in dfs:
    display(HTML(df.to_html()))
```

```python
#Write them all out to separate csv files
count = 0
for df in dfs:
    t= 't'+ str(count) + ".csv"
    df.to_csv(t)
    count +=1
```

```python
# Example at url below:
# https://xlsxwriter.readthedocs.io/example_pandas_multiple.html
!pwd
```

```python
writer = pd.ExcelWriter('big_tester.xlsx', engine='xlsxwriter')

i = 1
for df in dfs:
    sn = 'sheet' + str(i)
    df.to_excel(writer, sheet_name=sn)
    i+= 1
writer.save()    
```
